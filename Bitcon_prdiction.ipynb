```python
y=df.Close                                         ## Setting up Target 
X=df.drop("Close",axis=1)                          ## Droped target from dataframe
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler

# Create an instance of the StandardScaler class
sc = StandardScaler()

# Fit and transform the data using the scaler
X = sc.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20)           ## divding data in train test 80/20 ratio
```

```python
from sklearn.linear_model import LinearRegression

# Create an instance of LinearRegression
lr = LinearRegression()

# Fit the model using the training data
model = lr.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score,confusion_matrix,mean_absolute_error,mean_squared_error
```

```python
y_pred=model.predict(X_test)          ## model is prediting test data
```

```python
mean_absolute_error(y_test,y_pred)    # mean absolute error
```

```python
mean_squared_error(y_test,y_pred)     # Mean squared error
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))     # Root mean squared error
```

```python
model_accuracy=r2_score(y_test,y_pred)                 # cheacking r2 
```
